# Visualization Notebook
This notebook generates publication-ready charts from the processed scenario and battery analysis data.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

project_root = Path.cwd()
processed_dir = project_root / "data" / "processed"
visualizations_dir = project_root / "visualizations"
visualizations_dir.mkdir(parents=True, exist_ok=True)

scenarios_df = pd.read_csv(processed_dir / 'scenario_forecasts.csv')
scenarios_df['date'] = pd.to_datetime(scenarios_df['date'])
metrics_df = pd.read_csv(processed_dir / 'battery_storage_analysis.csv')
```

```python
plt.figure(figsize=(10, 5))
for scenario in ['worst', 'average', 'best']:
    subset = scenarios_df[scenarios_df['scenario'] == scenario]
    plt.plot(subset['date'], subset['renewable_total_mw'], label=scenario.capitalize())
plt.plot(scenarios_df[scenarios_df['scenario'] == 'average']['date'], scenarios_df[scenarios_df['scenario'] == 'average']['load_mw'], '--', color='black', label='Demand')
plt.title('Renewable Generation vs Demand by Scenario')
plt.xlabel('Date')
plt.ylabel('MW')
plt.legend()
plt.tight_layout()
plt.savefig(visualizations_dir / 'scenario_generation_vs_demand.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.barplot(data=metrics_df, x='battery_config', y='shortfall_reduction_pct', hue='scenario')
plt.title('Battery Shortfall Reduction by Configuration')
plt.xlabel('Battery Configuration')
plt.ylabel('Shortfall Reduction (%)')
plt.tight_layout()
plt.savefig(visualizations_dir / 'battery_shortfall_reduction.png', dpi=300)
plt.show()

print(f"Saved visualizations to {visualizations_dir}")
```
